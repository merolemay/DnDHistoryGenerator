```python
def cifrado_cesar(texto, clave):
    resultado = ""
    for letra in texto:
        if letra.isalpha():
            mayuscula = letra.isupper()
            letra = letra.lower()
            codigo = ord(letra)
            codigo += clave
            if codigo > ord('z'):
                codigo -= 26
            if codigo < ord('a'):
                codigo += 26
            letra_cifrada = chr(codigo)
            if mayuscula:
                letra_cifrada = letra_cifrada.upper()
            resultado += letra_cifrada
        else:
            resultado += letra
    return resultado

texto_original = "Hello World"
clave = 3
texto_cifrado = cifrado_cesar(texto_original, clave)
print(texto_cifrado)
```
