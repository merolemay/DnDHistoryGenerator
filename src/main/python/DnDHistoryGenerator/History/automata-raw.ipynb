```python
#q1 = peleas (1)
#q2 = te escondes (2)

#q3 = te acercas a la tribu (1)
#q4 = no te acercas a la tribu pero te ven y cedes ante la presion social (2)

#q5 = entras sigilosamente (1)
#q6 = entras estrepitosamente con la tribu, buscando pelea (2)

#q7 = liberas a xx e intentas huir (1) --> mueres (te matan los secuestradores)
#q8 = liberas a todos y huyes con xx (2) --> vives

#q9 = esperas a que baje el sol para seguir a la caravana secuestrada (1)
#q10 = sigues inmediatamente a la caravana secuestrada (2)

#q11 = sigues sigilosamente a la tribu (2)

#q12 = intentas liberar sigilosamente a xx tu solo (1) --> mueres (te mata la tribu creyendo que eras un espia de los secuestradores)
#q13 = intentas entrar estrepitosamente para liberar a xx tu solo (2) --> (se te une la tribu en ultimo momento, pierdes un brazo pero liberas a xx)

#  Given he input symbols {0, 1} and the language that accepts a string w that contains at least four 1â€™s, build a finite state automata
from pyformlang.finite_automaton import DeterministicFiniteAutomaton,State
input_symbols = {"1", "2"}
q0 = State("q0")
q1 = State("q1")
q2 = State("q2")
q3= State("q3")
q4 = State("q4")
q5 = State("q5")
q6 = State("q6")
q7= State("q7")
q8 = State("q8")
q9= State("q9")
q10 = State("q10")
q11 = State("q11")
q12 = State("q12")
q13 = State("q13")
states = {q0, q1, q2, q3, q4, q5, q6, q7, q8, q9, q10, q11,q12,q13}
initial_state = q0
final_states = {q7,q8, q12, q13}
transitions = [
    (q0, "1", q1),
    (q0, "2", q2),

    (q1, "1", q3),
    (q1, "2", q4),

    (q2, "1", q9),
    (q2, "2", q10),

    (q3, "1", q5),
    (q3, "2", q6),

    (q4, "1", q5),
    (q4, "2", q6),

    (q5, "1", q7),
    (q5, "2", q8),

    (q6, "1", q7),
    (q6, "2", q8),

    (q9, "1", q3),
    (q9, "2", q4),

    (q10, "1", q7),
    (q10, "2", q13),

    (q11, "1", q12),
    (q11, "2", q13)
]
automaton = DeterministicFiniteAutomaton(
    states=states ,
    input_symbols=input_symbols,
    start_state = initial_state,
    final_states = final_states)
automaton.add_transitions(transitions)
```
